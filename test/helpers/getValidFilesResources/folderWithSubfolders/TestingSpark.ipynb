## Este notebook comprueba que podemos conectar a Spark a tope de power

```python
# Importamos la librería
from pyspark import SparkContext

# Creamos un contexto de Spark
# El primer parámetro indica a qué instancia de Spark vamos a conectar. Local en nuestro caso
# El segundo es el nombre de la aplicación, podemos poner el que queramos
sc = SparkContext("local", "MyFirstApp")

# Parallelize crea un RDD (inmutable) con un array de elementos, en este caso strings
words = sc.parallelize (["vamos", "a", "probar", "que", "spark", "paraleliza", "guay"])

# Estos elementos se particionarán y se enviarán a los workers. En nuestro caso solo tenemos uno...

# Esto es simplemente para imprimirlo por pantalla
def f(x): print(x)
fore = words.foreach(f)
```
